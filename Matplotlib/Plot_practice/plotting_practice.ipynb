```python
import csv
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('practice_survey_data.csv', index_col='respondent_id') 

plt.style.use('ggplot')


years_coding=df['years_coding']
python_salary_usd=df['python_salary_usd']

plt.scatter(years_coding, python_salary_usd, edgecolor= 'black', linewidth=1.5, alpha=0.5)

plt.title('CODE EXPERIENCE SALARY')
plt.xlabel('Years_coding')
plt.ylabel('python_salary_usd')

plt.tight_layout()

plt.show()










```
