```python
#import from matplotlib library
from matplotlib import pyplot as plt
#Python dev data. 
age_x = [21,25,26,27,28,29,30,31,32, 35]#x axis
py_dev_y = [70000, 80000, 95000, 110000, 125000,
             140000, 155000, 175000, 190000,  210000,]#y axis

plt.plot(age_x, py_dev_y, color='k', linewidth='3', label='Ml eng')#plot the data (x axis, y axis, any color, any linewidth, label as a key for the legend )
#Javascript dev data. 
age_x = [21,25,26,27,28,29,30,31,32, 35]#x axis
js_dev_y = [65000, 75000, 85000, 95000, 110000,
             125000, 140000, 155000, 170000, 185000]# y axis

plt.plot(age_x, js_dev_y, color='orange', linewidth='3', label='js_dev')#plot the data (x axis, y axis, any color, any linewidth, label as a key for the legend )
#other devs data
age_x = [21,25,26,27,28,29,30,31,32, 35]#x axis
dev_y = [30000, 40000, 45000, 50000,60000, 62000, 65000, 70000, 72000,85000,]#y axis

plt.plot(age_x, dev_y, color='b', linestyle ='--', label='All Devs')#plot the data (x axis, y axis, any color, any linewidth, label as a key for the legend )

plt.xlabel('Ages')#label for the x axis
plt.ylabel('Median Salary(USD)')#label for the y axis
plt.title('Median Salary(USD) by Age')#label for the TITLE of the graph

plt.legend()# Legend to show the key and allocate each line belonging to which data
plt.grid('true')# for a grid
plt.tight_layout()# in case there an issue with padding 
plt.xkcd()# playful style/design of the graph
#plt.style.use(...) pickig a style/ design you want for the graph

plt.savefig('plot.png')#save the graph image to whatever directory you choose
plt.show()#show the plotted graph
```
