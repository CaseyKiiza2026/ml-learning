```python
sepal_length=df['sepal_length']
sepal_length_mean=df['sepal_length'].mean()

plt.hist(sepal_length, bins=10)
color='#fc4f30'

plt.axvline(sepal_length_mean, color=color, label='Sepal_length_ Median', linewidth= 2)

plt.legend()
plt.title('Sepal lenth stats')
plt.xlabel('Sepal length')
plt.ylabel('frequency')
plt.tight_layout()

plt.show()
plt.savefig('Iris_sepal_length_hist.png')
```
