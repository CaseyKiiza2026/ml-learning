```python
def youth(x):
    if x < 30:
         return 'Young'
    else:
         return 'Adult'

df['age_group']=df['AGE'].apply(youth)
df
```

```python
def incentives(x):
    if x >= 80000:
        return  10000
    else: 
        return  5000
         
    
df['bonus'] =df['SALARY'].apply(incentives)
df
```

```python
#name length
def length(x):
    return len(x)

df['name_length'] = df['FIRST_NAME'].apply(length)
df
```

```python
#Mapping country id
mapping= {
    'USA':1,
    'Canada':2,
    'UK':3
}

df['country_id']=df['COUNTRY'].map(mapping)
df
```

```python
#mapping nicknames
mapped= {
    'Corey': 'Core',
    'Jane': 'J',
    'John': 'Johnny',
    'Sarah': 'Sar',
    'Mike': 'Mikey',
    'Emma': 'Em'
}
df['nicknames']=df['FIRST_NAME'].map(mapped)
df
```
