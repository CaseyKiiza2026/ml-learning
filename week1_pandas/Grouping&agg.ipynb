```python
import pandas as pd
df = pd.read_csv('data/results.csv')
schema_df = pd.read_csv('data/schema.csv')
pd.set_option('display.max_column', 172)
pd.set_option('display.max_row', 85)
df.rename(columns={'ConvertedCompYearly':'Salary'}, inplace = True)
df.head(10)
```

```python
schema_df[schema_df['qname']=='ConvertedCompYearly']
```

```python
#Find the average salary for each country.
df.groupby('Country')
df.groupby('Country')['Salary'].mean()
```

```python
#Find the max salary for each country.

df.groupby('Country')['Salary'].max()
```

```python
#Find the min salary for each country.

df.groupby('Country')['Salary'].min()
```

```python
#Count how many respondents belong to each country
df.groupby('Country')['Salary'].count().head(15)
```

```python
#Find the total salary paid per country.
df.groupby('Country')['Salary'].sum().head(10)

```

```python
#These are the most important GroupBy questions left. Q9
#For each country, find: mean salary, max salary, min salary
df.groupby('Country')['Salary'].agg(['mean', 'max', 'min'])
```

```python
df.groupby('Country')['Salary'].agg(['mean', 'max', 'min', 'count'])
```

```python
#For each country, find:average Salary, average WorkExp
df.groupby('Country')[['Salary','WorkExp']].mean()
```
