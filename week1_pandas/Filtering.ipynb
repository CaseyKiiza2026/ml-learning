```python
import pandas as pd
```

```python

people = {
    "first": ["Corey", "Jane", "John", "Sarah", "Mike", "Emma"],
    "last": ["Schafer", "Doe", "Doe", "Smith", "Johnson", "Brown"],
    "age": [35, 28, 42, 19, 31, 24],
    "country": ["USA", "Canada", "USA", "UK", "Canada", "USA"],
    "salary": [85000, 65000, 95000, 30000, 72000, 55000]
}
```

```python
df = pd.DataFrame(people)
```

```python
#People from the USA
filt = df['country'] == 'USA'
#PEOPLE ABOVE 30
unc = df['age'] > 30
#people earning more than 70000
racks = df['salary'] > 70000
#people only from canada
can = df['country'] == 'Canada'
#people younger than 25
jeets = df['age'] < 25
#people  earning more than 60000
sixty_racks = df['salary']> 60000
#people younger than 30
Mille= df['age'] < 30
#people above 25 
boom = df['age'] >25 
#people below 40
alpha = df['age'] < 40
```

```python
df.loc[filt]
```

```python
#PEOPLE ABOVE 30
df.loc[unc]
```

```python
#people earning more than 70000
df.loc[racks]
```

```python
#people only from canada
df.loc[can]
```

```python
#people younger than 25
df.loc[jeets]
```

```python
#people from canada and older than 30
df.loc[can & unc]
```

```python
#people from USA and earning more than 60000
df.loc[filt & sixty_racks]
```

```python
#Show people from Canada AND earning more than 70000.
df.loc[can & racks]
```

```python
#Show people from USA AND younger than 30.
df.loc[filt & Mille]
```

```python
#Show people from USA OR Canada.
df.loc[filt | can]
```

```python
#Show people whose:age > 25, age < 40
df.loc[boom & alpha ]
```

```python
#Show people from:Canada, UK, using a method that doesn't require writing two separate country comparisons.
df.loc[~filt]
```

```python
#Show only:first country for people older than 30.
df.loc[ unc, ['first' , 'country']]
```

```python
#Exercise 17 Show only: first salary for people from Canada.
df.loc[can,['first','salary']]
```

```python
# Exercise 18 Show only: first age salary for people from USA earning more than 60000.
df.loc[filt & sixty_racks, ['first','age','salary']]
```
