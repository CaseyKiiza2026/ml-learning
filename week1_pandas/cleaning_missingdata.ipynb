```python
import pandas as pd
import numpy as np

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com',
              'JohnDoe@email.com', None, np.nan, None, 'Missing'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df=pd.DataFrame(people)
df.replace('Missing', np.nan, inplace=True)
df.replace('NA', np.nan, inplace=True)
```

```python
df.isna()
```

```python
#Display all missing values
df.fillna('Missing')
```

```python
#Count the number of missing values in each column.
df.isnull().sum()
```

```python
#Count the total number of missing values in the entire DataFrame.
df.isna().sum().sum()
```

```python
#Show only rows that contain at least one missing value.
df.isna().any()
```

```python
#Show only rows with no missing values.
df.isna().all(axis=0)
```

```python
#Replace all missing values with:
df.fillna('UNKNOWN')
```

```python
#replace missing emails with no email
df['email'] = df['email'].fillna('No Email')
```
